# FLIGHT FARE PREDICTION

## INTRODUCTION

In the fast-paced world of travel, finding the best deal on flight fares is a common challenge for both leisure and business travelers. The dynamic nature of airline pricing, influenced by factors such as demand, seasonality, and booking patterns, makes it difficult for passengers to anticipate and secure the most cost-effective options.

To address this issue, flight fare prediction has emerged as a valuable tool, leveraging data analysis, machine learning, and predictive modeling to forecast future airfares. This technology aims to empower travelers with insights that enable them to make informed decisions about when to book their flights, ultimately maximizing savings and optimizing their travel budgets.

Flight fare refers to the amount of money a passenger must pay to purchase a ticket for air travel on a commercial airline. It is the cost associated with a particular flight and includes charges for various services provided by the airline. The factors influencing flight fares can be complex and multifaceted, with several elements contributing to the overall cost. Common components of flight fares include:

<b>Base Fare:</b> <br><br>
    The fundamental cost of the flight ticket without any additional fees or taxes.</br></br>

<b>Taxes and Fees:</b><br><br>
Various taxes imposed by governments and additional fees charged by airports or regulatory authorities.</br></br>

<b>Fuel Surcharge:</b><br><br> A fee intended to cover the volatile costs of aviation fuel.</br></br>

<b>Airport Charges:</b><br><br> Charges imposed by airports for the use of facilities and services.</br></br>

<b>Service Class:</b><br><br> The type of seating or class of service chosen by the passenger (e.g., economy, business, first class), each with its associated fare.</br></br>

<b>Seasonal Factors:</b><br><br> Prices may vary based on the time of year, with peak travel seasons often associated with higher fares.</br></br>

<b>Demand and Supply:</b><br><br> Fares may be influenced by the demand for seats on a particular flight and the overall supply of available seats.</br></br>

<b>Booking Time:</b><br><br> The timing of booking can affect fares, with last-minute bookings often more expensive than those made well in advance.</br></br>

<b>Flight Duration and Stops:</b><br><br> Direct flights and shorter flight durations may be priced differently compared to flights with layovers or longer durations.</br></br>

<b>Airline Policies:</b><br><br> Different airlines may have varying pricing strategies based on factors such as brand positioning, competition, and marketing decisions.</br></br>

<b>Promotions and Discounts:</b><br><br> Airlines often run promotions, sales, or offer discounts, which can impact the final fare.</br></br>

Understanding the components of flight fare is crucial for travelers seeking the best deals and for airlines in managing pricing strategies. For data scientists working on flight fare prediction models, these factors become essential features to consider in the dataset, as they directly contribute to the overall cost of air travel for passengers.

In statistical terms, the first quartile (Q1) represents the value below which 25% of the data falls, and the third quartile (Q3) represents the value below which 75% of the data falls. These quartiles are useful for understanding the spread and distribution of the 'Price' variable in your dataset.








```python
Q1=data['Price'].quantile(0.25)
print("Q1 Quantile:",Q1)
Q3=data['Price'].quantile(0.75)
print("Q3 Quantile:",Q3)
```

- Q3 is the third quartile, which represents the value below which 75% of the data falls.
- Q1 is the first quartile, representing the value below which 25% of the data falls.
- IQR is the calculated Interquartile Range, which is the difference between Q3 and Q1.

```python
IQR = Q3-Q1
IQR
```

- lower_limit is calculated by subtracting 1.5 times the IQR from the first quartile (Q1).
- upper_limit is calculated by adding 1.5 times the IQR to the third quartile (Q3).
<br><br>
These limits are commonly used in identifying potential outliers in a dataset. Any value below the lower limit or above the upper limit is considered a potential outlier. Adjusting the multiplier (1.5 in this case) can be done based on the specific requirements of your analysis or the characteristics of your data. It's a way to define a range within which most of the data points are expected to lie and identify values that fall outside this range.</br></br>

```python
lower_limit = Q1-1.5*IQR
print("Lower limit:",lower_limit)
upper_limit = Q3+1.5*IQR 
print("Upper limit:",upper_limit)
```

This line of code creates a new DataFrame outliers_below_lower_limit that contains only the rows where the 'Price' values are below the calculated lower limit. These rows are considered potential outliers based on the lower limit determined using the Interquartile Range (IQR).

```python
data.loc[data['Price']<lower_limit]
```

This line of code creates a new DataFrame outliers_above_upper_limit that contains only the rows where the 'Price' values are above the calculated upper limit. These rows are considered potential outliers based on the upper limit determined using the Interquartile Range (IQR).

```python
data.loc[data['Price']>upper_limit]
```

<b>data['Price'] > upper_limit:</b> Creates a boolean mask indicating which rows have 'Price' values greater than the upper limit.

<b>data.loc[...]:</b> Uses boolean indexing to select rows where the condition is True.

<b>data['Price'].mean():</b> Calculates the mean (average) value of the 'Price' column.

The selected rows in the 'Price' column are then replaced with the calculated mean value.

This approach is one way to handle potential outliers by replacing them with a more central or representative value. Keep in mind that imputing outliers with the mean might affect the distribution and statistical properties of the data. Depending on your specific analysis and dataset, you might explore other techniques for handling outliers, such as removing them, transforming the data, or using more advanced imputation methods.

## USE TO LINEAR REGRESSION

- Linear regression is a statistical method used for modeling the relationship between a dependent variable and one or more independent variables by fitting a linear equation to observed data. The goal of linear regression is to find the best-fitting straight line (or hyperplane in higher dimensions) that minimizes the sum of squared differences between the observed values and the values predicted by the linear model.

- In simple linear regression, there is only one independent variable, while in multiple linear regression, there are multiple independent variables. The linear equation for simple linear regression can be expressed as:

- Y=β0+β1X+ϵ

where:

- Y is the dependent variable,
- X is the independent variable,
- β0 is the intercept (the value of Y when X is 0),
- β1 is the slope (the change in Y for a one-unit change in X),
- ϵ represents the error term, accounting for unobserved factors that affect Y but are not included in the model.

The goal is to estimate the values of β0 and β1 that minimize the sum of squared differences between the observed values and the values predicted by the linear equation.


Linear regression is widely used in various fields, including economics, finance, biology, and engineering, to analyze and model the relationships between variables.

```python
from sklearn.linear_model import LinearRegression
LR=LinearRegression()
LR.fit(x_train,y_train)
```

## USE TO KNN REGRESSION

- K-Nearest Neighbors (KNN) regression is a type of regression algorithm that utilizes the principles of the K-Nearest Neighbors algorithm for making predictions on continuous or numerical target variables. While traditional KNN is often associated with classification tasks, where it predicts the class of a data point based on the majority class of its k-nearest neighbors, KNN regression predicts a numerical value by averaging or taking a weighted average of the target values of its k-nearest neighbors.

<b>Here's a simplified explanation of how KNN regression works:</b>
<b>Neighbor Selection:</b><br><br> Given a new data point that needs a prediction, the algorithm identifies its k-nearest neighbors in the feature space. The "nearest" neighbors are determined based on a distance metric, commonly Euclidean distance, but other distance metrics can also be used</br></br>

<b>Prediction:</b><br><br> For regression, the predicted value for the new data point is typically the average (mean) of the target values of its k-nearest neighbors. In some cases, a weighted average may be used, giving more influence to closer neighbors

## USE TO DECISION TREE

- A decision tree is a popular supervised machine learning algorithm used for both classification and regression tasks. It is a tree-like model where an internal node represents a feature or attribute, the branches represent the decision rules, and each leaf node represents the outcome or prediction. Decision trees are used for making decisions based on input features by recursively splitting the dataset into subsets based on the values of the chosen features.
<br><br>
<b>Here are the key components of a decision tree:</b><br></br>

<b>Root Node:</b>

- The topmost node in the tree, representing the feature that provides the best split for the entire dataset. The feature is chosen based on criteria such as information gain (for classification) or reduction in variance (for regression).
<br>/br>
<b>Internal Nodes:</b></br></br>

- Nodes in the middle of the tree that represent features used for splitting the dataset. Each internal node corresponds to a decision or a test on a specific feature.
<br><br>
<b>Branches:</b></br></br>

- Branches connect nodes and represent the outcomes of the feature tests. Each branch corresponds to a possible value or range of values for the chosen feature.
<br><br>
<b>Leaves:</b></br></br>

- The terminal nodes of the tree, where predictions or classifications are made. Each leaf node represents the final output, whether it's a class label in classification or a numerical value in regression.


<b>Decision Rules:</b>
<br><br>
- The paths from the root node to the leaves represent decision rules based on the input features. These rules determine how the input features are evaluated to arrive at a final prediction.</br></br>

- Decision trees are constructed through a process called recursive binary splitting. At each internal node, the dataset is divided into two subsets based on the chosen feature and its threshold. This process is repeated recursively for each

## GRADIENT BOOSTING REGRESSOR

- Boosting regression refers to a machine learning technique that combines the predictions of multiple weak learners (typically simple models, often referred to as "base learners") to create a more accurate and robust predictive model. In the context of regression, boosting is used to improve the predictive performance of models that aim to predict a continuous outcome.

- The boosting algorithm works iteratively, where each weak learner is trained to correct the errors made by the previous ones. During each iteration, the algorithm assigns higher weights to the misclassified instances, emphasizing the importance of these instances in subsequent training rounds. The final prediction is then a weighted sum of the individual weak learner predictions.

- Popular boosting algorithms for regression include AdaBoost (Adaptive Boosting), Gradient Boosting, and XGBoost. These algorithms have been widely used in various applications due to their ability to handle complex relationships in the data and produce accurate predictions.

- Boosting is known for its ability to reduce bias and variance, leading to models with high predictive accuracy. However, it is essential to tune hyperparameters carefully to avoid overfitting, as boosting can be prone to fitting the training data too closely.

## USE TO RANDOM FOREST

- Random Forest is an ensemble learning method used for both classification and regression tasks. It operates by constructing a multitude of decision trees during training and outputs the mode (for classification) or mean prediction (for regression) of the individual trees.


<b>Here are the key components and characteristics of Random Forest:</b>


<b>Decision Trees:</b>

- Random Forest is built upon the foundation of decision trees. Decision trees are simple models that make decisions based on a series of conditions. They recursively split the data into subsets based on features to make predictions.

<b>Ensemble Learning:</b>

- Random Forest is an ensemble method, meaning it combines the predictions of multiple models to improve overall performance. In this case, it creates a forest of decision trees.

<b>Bootstrap Aggregating (Bagging):</b>

- The algorithm uses a technique called bagging, where multiple subsets of the training dataset are sampled with replacement (bootstrap samples). Each subset is then used to train an individual decision tree.

<b>Feature Randomization:</b>

- During the construction of each decision tree, a random subset of features is considered for splitting at each node. This introduces diversity among the trees and helps prevent overfitting.

<b>Voting (Classification) or Averaging (Regression):</b>

- For classification tasks, the mode (most frequent class) of the predictions from individual trees is taken as the final prediction. For regression tasks, the mean of the individual tree predictions is used.

<b>Robustness and Generalization:</b>

- Random Forests are known for their robustness and ability to generalize well to new, unseen data. They tend to handle noise and outliers better than individual decision trees.

<b>Hyperparameters:</b>

- Random Forest has hyperparameters that can be tuned to control its behavior, such as the number of trees in the forest, the maximum depth of each tree, and the number of features considered for each split.


Random Forest is widely used in practice due to its effectiveness, simplicity, and versatility.
It is applied in various domains, including finance, healthcare, and image analysis, among others. The robustness of Random Forest makes it a popular choice for many machine learning tasks.

## HYPERPARAMETER TUNNING TECHNIQUE

<b>Grid Search:</b>

Involves defining a grid of hyperparameter values and exhaustively searching through the grid. Time-consuming but can be effective for small hyperparameter spaces. Helps find the best combination of hyperparameters.

<b>Random Search:</b>

Randomly samples hyperparameter values from predefined ranges. More efficient than grid search when the hyperparameter space is large. Has a higher chance of finding good hyperparameter combinations.

<b>Bayesian Optimization:</b>

Uses probabilistic models to model the objective function and iteratively explores the hyperparameter space. Efficient for expensive black-box optimization functions. Balances exploration and exploitation to find the optimal set of hyperparameters.

<b>Genetic Algorithms:</b>

Inspired by natural selection, genetic algorithms involve evolving a population of potential solutions. Individuals (sets of hyperparameters) undergo selection, crossover, and mutation operations. Can be effective in searching for optimal hyperparameters.

<b>Gradient-Based Optimization:</b>

Involves using gradient information to update hyperparameters. Requires the objective function to be differentiable with respect to hyperparameters. May be used for specific models and optimization objectives.

<b>Ensemble Methods:</b>

Building ensembles of models with different hyperparameters and combining their predictions. Provides a robust approach, as ensembles can compensate for the weaknesses of individual models. Requires training multiple models with diverse hyperparameter settings.

<b>Automated Hyperparameter Tuning:</b>

Platforms like AutoML (Automated Machine Learning) tools can automate the hyperparameter tuning process. They use a combination of techniques to efficiently explore the hyperparameter space. Suitable for users without extensive machine learning expertise.

<b>Cross-Validation:</b>

Essential for hyperparameter tuning to ensure the model's performance is evaluated on different subsets of the data. Helps prevent overfitting to a specific training-validation split.

<b>Early Stopping:</b>

Monitoring the model's performance during training and stopping when it ceases to improve on a validation set. Prevents overfitting and reduces the risk of selecting suboptimal hyperparameters.

<b>Domain Knowledge:</b>

Leveraging domain knowledge to narrow down the hyperparameter search space. Understanding the characteristics of the data and problem can guide the selection of relevant hyperparameters.

## EVALUTATION MODEL TO PLOT

<b>Model Initialization:<b>

- This dictionary (models) contains instances of various regression models. The models include Linear Regression, K-Nearest Neighbors Regressor, Support Vector Regressor (SVR), Decision Tree Regressor, Bagging Regressor, Gradient Boosting Regressor, and Random Forest Regressor.


<b>Evaluation Metrics Initialization:</b>

- Three dictionaries (mean_squared_errors, mean_absolute_errors, r2_scores) are initialized to store the evaluation metrics for each regression model.


<b>Model Training and Evaluation Loop:</b>

- This loop iterates through each model, fits the model to the training data (x_train and y_train), predicts the target variable for the test data (x_test), and calculates three regression evaluation metrics: Mean Squared Error (MSE), Mean Absolute Error (MAE), and R-squared. The results are stored in the respective dictionaries.

<b>Plotting the Results:</b>

- Finally, the code creates a 3-subplot bar chart to visualize the performance of each model based on the three evaluation metrics. The first subplot represents Mean Squared Error, the second subplot represents Mean Absolute Error, and the third subplot represents R-squared. The chart is displayed using plt.show() from the Matplotlib library.

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

models = {
    'Linear_Regression': LinearRegression(),
    'KNeighborsRegressor': KNeighborsRegressor(),
    'SVR': SVR(),
    'Decision_Tree': DecisionTreeRegressor(),
    'Bagging_Regressor': BaggingRegressor(),
    'Gradient_Boost': GradientBoostingRegressor(),
    'Random_Forest': RandomForestRegressor()
}

mean_squared_errors = {}
mean_absolute_errors = {}
r2_scores = {}

for name, model in zip(models.keys(),models.values()):
    model.fit(x_train, y_train)
    y_predict = model.predict(x_test)
    mse = mean_squared_error(y_test, y_predict)
    mae = mean_absolute_error(y_test, y_predict)
    r2score = r2_score(y_test, y_predict)

    mean_squared_errors[name] = mse
    mean_absolute_errors[name] = mae
    r2_scores[name] = r2score

fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(10, 15))

ax1.bar(mean_squared_errors.keys(), mean_squared_errors.values(), color='k')
ax1.set_ylabel('Mean Squared Error')

ax2.bar(mean_absolute_errors.keys(), mean_absolute_errors.values(), color='r')
ax2.set_ylabel('Mean Absolute Error')

ax3.bar(r2_scores.keys(), r2_scores.values(), color='b')
ax3.set_ylabel('R-squared')

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
